### Data Cleaning and Exploration

```julia
using CSV
adv_stats = CSV.read("NBA_AdvStats.csv")
salaries = CSV.read("NBA_Salary.csv")
```

```julia
using DataFrames
adv_stats_df = DataFrame(adv_stats)
salaries_df = DataFrame(salaries)
```

```julia
colnames = ["Rank", "Player", "Team", "19_20_Salary", "20_21_Salary", "21_22_Salary", "22_23_Salary",
            "23_24_Salary", "24_25_Salary", "Signed_Using", "Total_Guaranteed"]
rename!(salaries_df, Symbol.(colnames))
```

```julia
# Split player name and ID into separate columns
players = split.(adv_stats_df.Player, '\\')
```

```julia
foreach(enumerate([:Player_Name, :Player_ID])) do (i, n)
           adv_stats_df[!, n] = getindex.(players, i)
end
```

```julia
adv_stats_df[:,[:Player, :Player_Name, :Player_ID]]
```

```julia
players = split.(salaries_df.Player, '\\')
foreach(enumerate([:Player_Name, :Player_ID])) do (i, n)
           salaries_df[!, n] = getindex.(players, i)
end
salaries_df[:,[:Player, :Player_Name, :Player_ID]]
```

```julia
# for players who played for multiple teams, identify the row for their total stats
adv_stats_df[!, :Remove] .= 0
for i = 2:nrow(adv_stats_df)
    if adv_stats_df[i, :Player_ID] == adv_stats_df[i-1, :Player_ID]
        adv_stats_df[i, :Remove] = 1
    end
end
```

```julia
# check the count of the rows to be removed
sum(adv_stats_df[!, :Remove])
```

```julia
# remove duplicate rows
adv_stats_df = adv_stats_df[adv_stats_df[!, :Remove] .< 1, :]
```

```julia
# clean positions for players with multiple listed
for i = 1:nrow(adv_stats_df)
    if adv_stats_df[i, :Pos] in ["PF-SF", "SF-SG", "SG-PF", "SG-SF", "PF-C"]
        adv_stats_df[i, :Pos] = SubString(adv_stats_df[i, :Pos], 1, 2)
    end
    if adv_stats_df[i, :Pos] in ["C-PF"]
        adv_stats_df[i, :Pos] = "C"
    end
end
```

```julia
unique(adv_stats_df.:Pos)
```

```julia
# sort salary df by player name
salaries_df = sort!(salaries_df, [:Player, Symbol("19_20_Salary")])
```

```julia
# combine rows for players with multiple (players get a separate row for each salary paid in 19/20)
salaries_df[!, :Remove] .= 0
for i = 1:nrow(salaries_df)-1
    if salaries_df[i, :Player_ID] == salaries_df[i+1, :Player_ID]
        salaries_df[i+1, Symbol("19_20_Salary")] = salaries_df[i+1, Symbol("19_20_Salary")] + salaries_df[i, Symbol("19_20_Salary")]
        salaries_df[i, :Remove] = 1
    end
end
```

```julia
# remove duplicate rows
salaries_df = salaries_df[salaries_df[!, :Remove] .< 1, :]
```

```julia
salaries_slim_df = salaries_df[:,[:Player_ID, Symbol("19_20_Salary")]]
```

```julia
players = join(adv_stats_df, salaries_slim_df, on = :Player_ID)
```
